## Analyzing Best-Sellers Trends Accross Product Categories

### Crosstab Analysis

```python
if "category" in df.columns and "isBestSeller" in df.columns:
 crosstab = pd.crosstab(df['category'], df['isBestSeller'], margins=True, normalize='index')
 print(crosstab)
```

```python
best_seller_proportion = crosstab[1].sort_values(ascending=False)
print(best_seller_proportion)
```

```python
## From these results, we can state that grocery, smart home security & lighting, as well as health and pesronal care have the highest proppotions os best-sellers. Things such as graphic vards or snowboard boots are not best-sellers
```

## Statistical Tests

```python
# crosstab MSZoning and SaleCondition
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result
```

```python
chi2_stats, chi2_pvalue, _, _ = chi2_contingency(crosstab_result)
print(f"Chi-square Test Statistic: {chi2_stats}")
print(f"P-value: {chi2_pvalue}")
```

```python
# The very high chi-square test statistic means there is a large difference between the observed and expevcted frequence in the crosstabl. Meaning it is most unlikely that between categroy and "isbestseller" is purely random. The P-value also indicates that the results are statistically significant
```

```python
n = crosstab_result.sum().sum()
cramer_v = np.sqrt(chi2_stats / (n * (min(crosstab_result.shape) - 1)))
print(f"Cramér's V: {cramer_v:.2f}")
```

```python
# The Cramer's V being so low meaning there is a weak association between product cateogry and best-seller status.
```

```python
# This means that while product category does have an influence on whether a product becomes a best-seller but it is not a dominant factor
```
